** Exact Process Flow (for your case)

Summary Statistics: Compute mean, standard deviation, min, max for each treatment × parameter.

Data Cleaning: Ensure treatments and block identifiers are consistent.

Block Creation: Combine Run × Rep → Block.

Diagnostics: Check assumptions (normality, variance homogeneity).

Transform if needed: arcsine√ for % injury, √ for skewed biometrics.

Fit RCBD ANOVA: response ~ Treatment + Block.

Check ANOVA table: Is Treatment significant?

Post‑hoc (Tukey HSD): Compare treatments.

CLD grouping: Assign letters for easy interpretation.

Back‑transform means: Report in natural units (%, g, cm).

Final reporting: Means ± SE + CLD letters in tables **

```python
import pandas as pd

# === USER SETTINGS ===
# Change only these two lines:
file_name = "SA.xlsx"            # <-- put your Excel file name here
parameter = "Injurylevel_4WAS"        # <-- choose one: "Injurylevel_2WAS", "Injurylevel_4WAS", "Freshweight_4WAS", "Length_4WAS"

# === Load and clean data ===
df = pd.read_excel(file_name)
df.columns = df.columns.str.strip()
df["Treatment"] = df["Treatment"].astype(str).str.strip()

# === Summary statistics by Treatment ===
summary = df.groupby("Treatment")[parameter].agg(
    n="count",
    mean="mean",
    std="std",
    min="min",
    max="max"
).reset_index()

# === Print results ===
print(f"\nSummary statistics for {parameter}:\n")
print(summary.to_string(index=False))

# === Save results to Excel and CSV ===
output_excel = f"Summary_{parameter}.xlsx"
output_csv   = f"Summary_{parameter}.csv"

summary.to_excel(output_excel, index=False)
summary.to_csv(output_csv, index=False)

print(f"\nFiles saved: {output_excel} and {output_csv}")
```

```python

import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt

# === USER SETTINGS ===
file_name = "SA_FW.xlsx"   # <-- change to your dataset file name

# === Load and clean data ===
df = pd.read_excel(file_name)
df.columns = df.columns.str.strip()
df["Treatment"] = df["Treatment"].astype(str).str.strip()
df["Run"] = df["Run"].astype(str).str.strip()
df["Rep"] = df["Rep"].astype(str).str.strip()
df["Block"] = df["Run"] + "_Rep" + df["Rep"]

# === Variables to test (raw data) ===
variables = {
    # "Injurylevel_2WAS": "Injury at 2 weeks (%)",
    # "Injurylevel_4WAS": "Injury at 4 weeks (%)",
    "Freshweight_4WAS": "Fresh weight (g)",
    # "Length_4WAS": "Length (cm)"
}

results = []

for var, label in variables.items():
    print(f"\n=== Diagnostics for {label} ===")

    # Fit RCBD ANOVA model
    model = smf.ols(f"{var} ~ C(Treatment) + C(Block)", data=df).fit()
    residuals = model.resid

    # Shapiro–Wilk test
    shapiro_p = shapiro(residuals)[1]

    # Levene’s test (variance across treatments)
    groups = [df[var][df["Treatment"] == t].dropna() for t in df["Treatment"].unique()]
    levene_p = levene(*groups)[1]

    # Flags
    norm_flag = "✅ Normal" if shapiro_p > 0.05 else "❌ Not normal"
    var_flag  = "✅ Homogeneous" if levene_p > 0.05 else "❌ Variances unequal"

    print(f"Shapiro–Wilk p-value: {shapiro_p:.4f} → {norm_flag}")
    print(f"Levene’s test p-value: {levene_p:.4f} → {var_flag}")

    results.append({
        "Variable": label,
        "Shapiro_p": round(shapiro_p,4),
        "Normality": norm_flag,
        "Levene_p": round(levene_p,4),
        "Variance": var_flag
    })

# === Create summary table as image ===
summary_df = pd.DataFrame(results)

fig, ax = plt.subplots(figsize=(8,2))
ax.axis('off')
table = ax.table(cellText=summary_df.values,
                 colLabels=summary_df.columns,
                 cellLoc='center',
                 loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
plt.title("Diagnostics Summary (Raw Data)", fontsize=14, pad=20)
plt.show()

# === Save table to Excel for reviewer ===
summary_df.to_excel("Diagnostics_RawData.xlsx", index=False)
print("\nDiagnostics summary saved to Diagnostics_RawData.xlsx")
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt
import seaborn as sns

# === USER SETTINGS ===
file_name = "SA_FW.xlsx"   # <-- Excel file name

# === Load and clean data ===
df = pd.read_excel(file_name)
df.columns = df.columns.str.strip()
df["Treatment"] = df["Treatment"].astype(str).str.strip()
df["Run"] = df["Run"].astype(str).str.strip()
df["Rep"] = df["Rep"].astype(str).str.strip()
df["Block"] = df["Run"] + "_Rep" + df["Rep"]

# === Variables and transformations ===
# Define which transformation to apply for each variable
variables = {
    # "Injurylevel_2WAS": ("Injury at 2 weeks (%)", "arcsine"),   # arcsine√ recommended for % data
    # "Injurylevel_4WAS": ("Injury at 4 weeks (%)", "arcsine"),      # raw only (transformation ineffective)
    "Freshweight_4WAS": ("Fresh weight (g)", "sqrt"),           # √ transformation needed
    # "Length_4WAS": ("Length (cm)", "none")                      # raw acceptable
}

results = []

for var, (label, transform) in variables.items():
    print(f"\n=== Diagnostics for {label} ({transform}) ===")

    # Apply transformation
    if transform == "arcsine":
        df["response"] = np.arcsin(np.sqrt(df[var]/100.0))
    elif transform == "sqrt":
        df["response"] = np.sqrt(df[var] + 0.5)
    else:
        df["response"] = df[var]

    # Fit RCBD ANOVA model
    model = smf.ols("response ~ C(Treatment) + C(Block)", data=df).fit()
    residuals = model.resid

    # Shapiro–Wilk test
    shapiro_p = shapiro(residuals)[1]

    # Levene’s test (variance across treatments)
    groups = [df["response"][df["Treatment"] == t].dropna() for t in df["Treatment"].unique()]
    levene_p = levene(*groups)[1]

    # Flags
    norm_flag = "✅ Normal" if shapiro_p > 0.05 else "❌ Not normal"
    var_flag  = "✅ Homogeneous" if levene_p > 0.05 else "❌ Variances unequal"

    print(f"Shapiro–Wilk p-value: {shapiro_p:.4f} → {norm_flag}")
    print(f"Levene’s test p-value: {levene_p:.4f} → {var_flag}")

    results.append({
        "Variable": label,
        "Transform": transform,
        "Shapiro_p": round(shapiro_p,4),
        "Normality": norm_flag,
        "Levene_p": round(levene_p,4),
        "Variance": var_flag
    })

    # Diagnostic plots
    plt.figure(figsize=(15,4))
    plt.suptitle(f"Residual diagnostics for {label} ({transform})", fontsize=14)

    # Histogram
    plt.subplot(1,3,1)
    sns.histplot(residuals, kde=True)
    plt.title("Residual Histogram")

    # Q-Q plot
    plt.subplot(1,3,2)
    sm.qqplot(residuals, line='s', ax=plt.gca())
    plt.title("Q-Q Plot")

    # Residuals vs Fitted
    plt.subplot(1,3,3)
    plt.scatter(model.fittedvalues, residuals)
    plt.axhline(0, color='red', linestyle='--')
    plt.xlabel("Fitted values")
    plt.ylabel("Residuals")
    plt.title("Residuals vs Fitted")

    plt.tight_layout()
    plt.show()

# === Summary table ===
summary_df = pd.DataFrame(results)
print("\n=== Diagnostic summary after transformations ===")
print(summary_df.to_string(index=False))
```
